```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def polar(R, a):
    x = R * np.cos(a)
    y = R * np.sin(a)
    return x, y
```

```python
t = np.arange(-2 * np.pi, 2 * np.pi, 0.01)
r = 1
x = r * np.cos(t)
y = r * np.sin(t)

plt.polar(x, y)
plt.show()
```

1. Решите систему уравнений: 
$$ \left\{
\begin{array}{l}
y = x^2 – 1 \\

1. Решите систему уравнений и неравенств:
$$\left \{
\begin{array}{l}
y = x^2 – 1 \\
exp(x) + x∙(1 – y) > 1
\end{array}
\right .$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-3, 3, 200)
y = x**2 - 1

plt.plot(x, y, label=f'$y = x^2 – 1$')

y = (x - 1 + np.exp(x)) / x
plt.plot(x, y, label=f'$exp(x) + x∙(1 – y) = 1$')

plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
from scipy.optimize import fsolve
```

### 1

```python
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x *(1-y) -1)

print(f'x_1:{fsolve(equations, (-2, 1))[0]} y_1:{fsolve(equations, (-2, 1))[1]}\n',
      f'x_2:{fsolve(equations, (2, 4))[0]} y_2:{fsolve(equations, (2, 4))[1]}')
```

### 2

```python
print(f'x_1:{fsolve(equations, (-2, 1))[0]} y_1:{fsolve(equations, (-2, 1))[1]}\n',
      f'x_2:{fsolve(equations, (0, 1.5))[0]} y_2:{fsolve(equations, (0, 1.5))[1]}\n'
      f'x_3:{fsolve(equations, (2, 4))[0]} y_2:{fsolve(equations, (2, 4))[1]}')
```

```python
x1, y1 = fsolve(equations, (-2, 1))
x2, y2 = fsolve(equations, (0, 1.5))
x3, y3 = fsolve(equations, (2, 4))
print(x1, y1)
print(x2, y2)
print(x3, y3)
```
